## Heatmap Correlation

Because this dataset about weather related disease, we need to see is that the disase has the correlation or not, so in this procces we make a visualisation with heatmap correlation, since the dataset have so many feature this visualisation can help us to know the total values.

```python
# make a correlation heatmap
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
plt.figure(figsize=(14, 8))
# show theh correlation heatmap only with values greater than 0.5
plt.title('Correlation Heatmap')
sns.heatmap(df.corr(), mask=mask, cmap='coolwarm', annot=False, fmt='.2f', linewidths=.5)

plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Filter the correlation matrix to show only correlations greater than 0.5
filtered_correlation = correlation_matrix[correlation_matrix > 0.5]

# Create a heatmap for the filtered correlation matrix
plt.figure(figsize=(18, 15))
sns.heatmap(filtered_correlation, cmap='coolwarm', annot=False, fmt='.2f', linewidths=.5, mask=mask)
plt.title('Heatmap of Features with Correlation > 0.5')
plt.show()
```

```python
# we gonna take the corr values greater than 0.5 and less than 1
high_corr = correlation_matrix[(correlation_matrix > 0.5) & (correlation_matrix < 1.0)]

# get the coloumn names of the high correlation values
high_corr_columns = high_corr.columns[high_corr.any()].tolist()
print("Columns with high correlation values:")
print(high_corr_columns)

```

## Age, Tempeature, Humidity, and Wind Speed Distibution

```python
# make a subplot for distribution Age, Temperature, Humidity, and WindSpeed
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Distribution of Age, Temperature, Humidity, and WindSpeed')

#histogram Age
sns.histplot(df['Age'], bins=20, kde=True, ax=axes[0, 0])
# add mean lines
axes[0, 0].axvline(df['Age'].mean(), color='r', linestyle='--', label='Mean')
# add legend
axes[0, 0].legend()
axes[0, 0].set_title('Age Distribution') # set title for the first subplot

# histogram Temperature
sns.histplot(df['Temperature (C)'], bins=20, kde=True, ax=axes[0, 1])
# add mean lines
axes[0, 1].axvline(df['Temperature (C)'].mean(), color='r', linestyle='--', label='Mean')
# add legend
axes[0, 1].legend()
axes[0, 1].set_title('Temperature Distribution') # set title for the second subplot

# histogram Humidity
sns.histplot(df['Humidity'], bins=20, kde=True, ax=axes[1, 0])
# add mean lines
axes[1, 0].axvline(df['Humidity'].mean(), color='r', linestyle='--', label='Mean')
# add legend
axes[1, 0].legend()
axes[1, 0].set_title('Humidity Distribution')  # set title for the third subplot

# histogram Wind Speed
sns.histplot(df['Wind Speed (km/h)'], bins=20, kde=True, ax=axes[1, 1])
# add mean lines
axes[1, 1].axvline(df['Wind Speed (km/h)'].mean(), color='r', linestyle='--', label='Mean')
# add legend
axes[1, 1].legend()
axes[1, 1].set_title('WindSpeed Distribution') # set title for the fourth subplot

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

As we can see, **Age** feature has even frequent of values among the other feature. We also realise that feature like **Temperature**, **Humidity**, and **Wind Speed** has **skewed distribution**, but we don't need to fix that kind of feature, because this feature are real data that happen on outside, dropping the outlier is not a good decision and might reduce the accuracy and sensitivity of the model later. 

## Deatiled Age Distribution and Value Counts

```python
# Setup layout with GridSpec
fig = plt.figure(figsize=(14, 10))

# we use gridspec to make a subplot with 2 rows and 2 columns in 1 row
# because in subplot we cannot make a subplot with 2 rows and 2 columns in 1 row with different height
# with gridspec more flexible to make a subplot with different height
gs = fig.add_gridspec(2, 2, height_ratios=[1, 0.6])  # 2 rows, 2 columns
fig.suptitle('Age Distribution and Value Count')


# Subplot for age distribution by gender = 0
ax1 = fig.add_subplot(gs[0, 0])
sns.histplot(df[df['Gender'] == 0]['Age'], bins=20, kde=True, ax=ax1, color='steelblue')
# add mean lines
ax1.axvline(df[df['Gender'] == 0]['Age'].mean(), color='r', linestyle='--', label='Mean')
# add legend
ax1.legend()
# add title
ax1.set_title('Age Distribution - Gender 0')

# Subplot for age distribution by gender = 1
ax2 = fig.add_subplot(gs[0, 1])
sns.histplot(df[df['Gender'] == 1]['Age'], bins=20, kde=True, ax=ax2, color='salmon')
# add mean lines
ax2.axvline(df[df['Gender'] == 0]['Age'].mean(), color='r', linestyle='--', label='Mean')
# add legend
ax2.legend()
# add title
ax2.set_title('Age Distribution - Gender 1')

# Subplot for countplot gender
ax3 = fig.add_subplot(gs[1, :])  # ":" means we take 2 columns in 1 row
sns.countplot(data=df, x='Gender', palette='mako', ax=ax3)
ax3.set_title('Count of Each Gender')

plt.tight_layout()
plt.show()
```

We move on to see the details of the **Age** by **Gender** feature. If we assume that the gender value of 0 is male and 1 is female, then we know that the frequency of a female has more than a male. Furthermore, both genders show the same values of mean and that is between 40 and 60. The male gender has a high frequency of age 50, and the female gender has a high frequency of age 60 and 80.

## Progonosis Value Counts

```python
# make barplot for the prognosis value counts
plt.figure(figsize=(14, 6))

order =  df['prognosis'].value_counts().index  # Get the order of prognosis values based on their counts

plt.title('Prognosis Value Counts')
sns.countplot(data=df, y='prognosis', palette='mako', order=order)  # Use the order for the x-axis
plt.xlabel('Prognosis')
plt.ylabel('Count')
plt.show()
```

**Prognosis** is a long term prediction to the disease and focus on the progress of how we respond to the disease. The difference between prognosis and diagnosis is that diagnosis is focused on the disease and we are judging the disease, but prognosis is focused on the progress of the disease. So these data show the prognosis of a lot of the diagnosis that we have and prognosis is the prediction of how the disease will progress in the future. We know from this data that **Heart Attack**, **Migraine** and **Influenza** are the **top 3** predictions of how the disease will come. So in the next process we're going to make a model to predict the prognosis and we're also going to see which disease has more influence.

## Progonosis Value Counts by Gender

```python
plt.figure(figsize=(14, 6))

order =  df['prognosis'].value_counts().index  # Get the order of prognosis values based on their counts

plt.title('Prognosis Value Counts by Gender')
sns.countplot(data=df, y='prognosis', palette='mako', hue= 'Gender', order=order)  # Use the order for the x-axis
plt.xlabel('Prognosis')
plt.ylabel('Count')
plt.show()
```

If we look at the prognosis based on **Gender** feature, from the top 3 our prognosis the frequency of Heart Attack has more Female than a male, frequency of Migraine has more male than female, and Influenza has more female than a male. But overall the prognosis that we have a lot of experience on female.
